```
# Excercise 29
# https://www.practicepython.org/exercise/2016/08/03/29-tic-tac-toe-game.html

 
def convertNumtoLetter(num):
  if num == 0:
    return " "
  if num == 1:
    return "X"
  if num == 2:
    return "O"

def pHorizontal():
  print(' ---' * 3)

def pVertical(gameRow):
  for i in range(3):
    print('| {} '.format(convertNumtoLetter(gameRow[i])), end = '')
  print('|')

def drawBoard(gameState):

  for i in [0,1,2]:
    pHorizontal()
    pVertical(gameState[i])
  pHorizontal()

def getPlayerMove(playerTurn):
  return input('It is Player {0}\'s Turn! \nMake a move in the form [row,col] ! \n'.format(playerTurn))

def validatePlayerMove(userMove):
  userMoveRow = int(userMove[0])
  userMoveCol = int(userMove[1])

  if userMoveRow > 1 or userMoveRow < 3:

    if userMoveCol > 1 or userMoveCol < 3:

      return [userMoveRow, userMoveCol]

def makeMove(gameState, currentPlayer, playerMove):
  gameState[playerMove[0]][playerMove[1]] = currentPlayer
  return gameState

def checkVictory(gameState):

  # rows
	for i in range(0,3):
		row = set([gameState[i][0],gameState[i][1],gameState[i][2]])
		if len(row) == 1 and gameState[i][0] != 0:
			return gameState[i][0]

	# columns
	for i in range(0,3):
		column = set([gameState[0][i],gameState[1][i],gameState[2][i]])
		if len(column) == 1 and gameState[0][i] != 0:
			return gameState[0][i]

	# diagonals
	diag1 = set([gameState[0][0],gameState[1][1],gameState[2][2]])
	diag2 = set([gameState[0][2],gameState[1][1],gameState[2][0]])
	if len(diag1) == 1 or len(diag2) == 1 and gameState[1][1] != 0:
		return gameState[1][1]

	return 0

def checkMaxTurns(moveCounter):
  if moveCounter >= 9:
    return 3
  else:
    return 0

def playGame():

  # Game Loop
  while True:

    gameFinished = 0

    # Win array
    # 0 - Draws
    # 1 - P1 
    # 2 - P2
    wins = [0, 0, 0]

    moveCounter = 0

    gameState = [[0, 0, 0],
                 [0, 0, 0],
                 [0, 0, 0]]

    while gameFinished == 0:

      drawBoard(gameState)
      
      if moveCounter % 2 == 0:
        currentPlayer = 1
        playerMove = getPlayerMove(currentPlayer)
        
      else:
        currentPlayer = 2
        playerMove = getPlayerMove(currentPlayer)

      # Validate Player Move
      vMove = validatePlayerMove(playerMove.split(","))
      print('Move Validated {}'.format(vMove))

      # Make Player Move
      gameState = makeMove(gameState, currentPlayer, vMove)

      # Check Victory
      gameFinished = checkVictory(gameState)
      # Check End Condition
      gameFinished = checkMaxTurns(moveCounter)

      # Increment Move Counter
      moveCounter += 1

    if gameFinished == 1:
      print('Player 1 WINS!')
      wins[1] +=1
    elif gameFinished == 2:
      print('Player 2 WINS!')
      wins[2] +=1
    elif gameFinished == 3:
      print('DRAW! No one Wins!')
      wins[0] +=1

  print('Current Statistics\nDraws - {0}\nPlayer 1 - {1}\nPlayer 2 - {2}\n'.format(wins[0], wins[1], wins[2]))
  userOption = input('Press 0 to play again, Press any other key to exit')
  if int(userOption) != 0:
    return 0

playGame()


```

```
# Excercise 9
# https://www.practicepython.org/exercise/2014/04/02/09-guessing-game-one.html

import random

def playGame():

  target = random.randint(1,9)

  guessCount = 0

  while True:
    playerInput = input('What is your Guess?\n')
    if playerInput == 'exit':
      print('Exiting...')
      return 0
    guess = int(playerInput)

    guessCount += 1

    if target == guess:
      print('WOO! You got it right!\nIt only took {0} guess(es)'.format(guessCount))
      return 0
    elif target > guess:
      print('Too Low :(')
    elif target < guess:
      print('Too High :(')


playGame()
```
