```
# Excercise 11
# https://www.practicepython.org/exercise/2014/04/16/11-check-primality-functions.html


def checkDivisors(checkNum):

  possNums = range(1, checkNum)
  return [ x for x in possNums if checkNum % x == 0]


checkNum = int(input('Input a number to check primes: '))

divisors = checkDivisors(checkNum)
if len(divisors) > 1:
  print('{0} is not prime!\nThese are its divisors\n{1}'.format(checkNum, divisors))
else:
  print('OMG! {0} is prime!\nIts only divisor is {1}'.format(checkNum, divisors))

```

```
# Excercise 18
# https://www.practicepython.org/exercise/2014/07/05/18-cows-and-bulls.html

import random


def getAndValidateInput():
  validInput = False

  while ~validInput:

    playerInput = input('Make your Guess! : ')

    # Convert to integer
    try:
      intGuess = int(playerInput)
    except:
      print('Not an integer| Guess in the range 1000-9999:/ \n Guess again')
    else:

      # Check length

      if intGuess > 9999:
        print('Too long! Guess in the range 1000-9999:/ \n Guess again')
      elif intGuess < 1000:
        print('Too short! Guess in the range 1000-9999:/ \n Guess again')
      else:
        return intGuess

def cows(target, guess):

  cowsCount = 0

  for i in range(len(target)):

    if target[i] == guess[i]:
      cowsCount += 1

  return cowsCount


def bulls(target, guess):

  bullsCount = 0

  for i in target:
    for j in guess:

      if i == j:
        bullsCount += 1
        break # not usually a fan of a break but it may have to do

  return bullsCount


def checkGameState(target, guess):

  # Cows - Correct Number, Correct Place
  # Bulls - Correct Number, Incorrect Place

  print([target[0], guess])

  targetArr = [int(i) for i in str(target[0])]
  guessArr = [int(i) for i in str(guess)]

  print([targetArr, guessArr])

  cowsCount = cows(targetArr, guessArr)
  bullsCount = bulls(targetArr, guessArr)

  return [cowsCount, bullsCount]




def cowsAndBulls():
  
  target = random.sample(range(1000, 9999),1)

  gameFinished = False
  guessCount = 0

  print('Cows and Bulls Guessing Game!\n I have a random 4 digit number in mind')

  while gameFinished == False:

    guess = getAndValidateInput()

    guessCount += 1

    cAndB = checkGameState(target, guess)

    if cAndB[0] == 4:

      gameFinished = True

      print('WOO! You guessed {0} correctly!\n It only took {1} guesses'.format(target, guessCount))

    else:

      print('Cows [{0}] Bulls [{1}] Guess [{2}]'.format(cAndB[0],cAndB[1],guess))




cowsAndBulls()
```

```
# Excercise 25
# https://www.practicepython.org/exercise/2015/11/01/25-guessing-game-two.html



# modified binary search algorithm to take user feedback
def guesserAIThatWillTakeOverAllTheWorld(): 
  # lowest number in range
  low = 0
  # highest number in range
  high = 100
  # midpoint
  m = int(high / 2)

  # number of guess attempts
  guessCount = 0

  # User Input Guide
  # - Guess too high / Lower
  # = Guess Correct / Spot on
  # + Guess too low / Higher

  print('Think of a number\n+ : higher\n- : lower\n= : spot on!')
  
  state = input('Is your number {}? '.format(m))

  while state != "=":
    guessCount += 1

    if state == "+":
      low = m + 1

    elif state == "-":
      high = m - 1

    # Lazy Rounding using int type cast
    # Recaulate middle point
    m = int((low + high) / 2)

    state = input('Is your number {}? '.format(m))

  print('Gotcha! It took {} guesses'.format(guessCount))

guesserAIThatWillTakeOverAllTheWorld()
```

```
# Excercise 16
# https://www.practicepython.org/exercise/2014/05/28/16-password-generator.html

import random
import string

def split(word): 
    return [char for char in word]

def setAlphabetStrength(strength):

  lowerCase = list(string.ascii_lowercase)
  upperCase = list(string.ascii_uppercase)
  numbers = list(string.digits)
  symbols = list(string.punctuation)

  alphabet = []

  if strength == 1:
    alphabet = lowerCase
  elif strength == 2:
    alphabet = lowerCase + upperCase
  elif strength == 3:
    alphabet = lowerCase + upperCase + numbers
  elif strength ==4:
    alphabet = lowerCase + upperCase + numbers + symbols

  return alphabet



def generatePassword(length = 10, strength = 3):
  # Length - desired pass output
  # strength - strength setting

  alphabet = setAlphabetStrength(strength)

  password = []

  for i in range(length):

    charIndex = random.randint(0, len(alphabet)-1)

    password += str(alphabet[charIndex])


  return ''.join(password)

def password1Line(length = 10, chars=string.ascii_letters + string.digits + string.punctuation):
	return ''.join(random.choice(chars) for _ in range(length))

length = 16
strength = 4

print([generatePassword(length, strength),password1Line(length)])
```
